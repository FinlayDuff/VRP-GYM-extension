# Import the packages and modules

```python
from gym_vrp.envs import VRPEnv, IRPEnv, TSPEnv, SantaIRPEnv
from agents import TSPAgent, IRPAgent, VRPAgent

import torch
```

## Train the agent

```python
batch_size = 10 # 256
seed = 23
num_nodes = 5 # 20
```

```python
env_tsp = TSPEnv(num_nodes=num_nodes, batch_size=batch_size, seed=seed)

agent_tsp = TSPAgent(
    seed=seed, csv_path=f"./train_logs/loss_log_tsp_{num_nodes}_{seed}.csv",
)
agent_tsp.train(
    env_tsp,
    epochs=251,
    check_point_dir=f"./check_points/tsp_{num_nodes}_{seed}/",
        )
```

## Visualise the actions of the agent in the environments

```python
env = TSPEnv(num_nodes=num_nodes, batch_size=batch_size, seed=seed, num_draw=3)
TSPModel='./check_points/tsp_5_23/model_epoch_250.pt'
```
